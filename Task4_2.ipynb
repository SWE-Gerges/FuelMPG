```python
#importing
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Keras and Tensorflow
import tensorflow.python.keras
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
```

```python
data = pd.read_csv('autoMPG.csv')
data_CPY = data
```

```python
data['Horsepower'].fillna(value = data['Horsepower'].mean(), inplace = True)
```

```python
data = data.drop('CarName', axis = 1)
```

```python
data.columns
```

```python
sc = StandardScaler()
```

```python
y = data['MPG']
X = data.drop('MPG', axis = 1)
```

```python
X.columns
```

```python
X = pd.DataFrame(sc.fit_transform(X), columns = ['Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration',
       'Model Year', 'Origin'])
#y = pd.DataFrame(sc.fit_transform(y), columns = ['MPG'])
```

```python
print(y)
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(X_train)
print(y_train)
```

```python
X.isna().any()
```

```python
model = Sequential()
```

```python
model.add(Dense(10, input_dim = 7, activation = 'relu'))
model.add(Dense(5, activation = 'relu'))
model.add(Dense(1))
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
history = model.fit(X_train, y_train, epochs = 500, batch_size = 64)
```

```python
y_pred = model.predict(X_test)
```
